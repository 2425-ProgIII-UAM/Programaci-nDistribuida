## Ejercicio 2: El torneo de Sudokus (3 puntos)
En el torneo de sudokus hay normas muy estrictas. Hay dos jugadores por mesa. El primer jugador que llega a la mesa se sienta en la silla de la izquierda, y el segundo en la de la derecha. Además, los jugadores no pueden darle la vuelta al papel con el sudoku a resolver, situado en el centro de la mesa, hasta que los dos jugadores se hayan sentado. Como los jugadores van despistados resolviendo sudokus en sus teléfonos, a veces no se dan cuenta y empiezan a resolver el sudoku antes de que llegue su contrincante. ¡Incluso hay jugadores que se han sentado encima del otro en la primera silla, porque cuando miraron estaba vacía!.

### Preguntas
- (2 puntos) Simulando una de las mesas con sus jugadores usando hilos o procesos, usa herramientas de la programación concurrente para evitar que estas dos situaciones incómodas y descalificantes ocurran. Indica los cambios necesarios en el siguiente código y justifícalos.
- (1 puntos) Se han dado cuenta en el torneo que a veces el jugador que tarda más en resolver el sudoku es el que acaba proclamado vencedor. Indica cómo resolverías la situación.



```python
from threading import Thread
import random

sillaIzquierdaOcupada = False
sudokuBocabajo = True

# variable que guarda quien ha ganado 
# mientras vala 0, nadie ha ganado
resueltoPorJugador = 0


def sentarse(posición="Izquierda"):
    global sillaIzquierdaOcupada
    
    if posición == "Izquierda":
        sillaIzquierdaOcupada = True

def darLaVueltaAlSudoku():
    global sudokuBocabajo
    sudokuBocabajo = False

def mirarSudoku(jugador):
    print ("El jugador ", jugador, "mira el sudoku")

def resolverSudoku(jugador=0):
    # hay un 98% de posibildades que en cada paso, el jugador tenga que seguir
    # pensando antes de resolver el sudoku
    while (resueltoPorJugador == 0 and random.random() < 0.99):
        if not sudokuBocabajo:
            mirarSudoku(jugador)
        print("Jugador", jugador, "Pensando...")
    print("Resuelto por jugador", jugador,"!")

def jugarPartida(jugador = 0):
    global resueltoPorJugador
    
    if not sillaIzquierdaOcupada:
        sentarse(posición="Izquierda")
    else:
        sentarse(posición="Derecha")

    if sudokuBocabajo:
        darLaVueltaAlSudoku()

    resolverSudoku(jugador)
    resueltoPorJugador = jugador

jugador1 = Thread(target=jugarPartida, kwargs={"jugador":1})
jugador2 = Thread(target=jugarPartida, kwargs={"jugador":2})

jugador1.start()
jugador2.start()

jugador1.join()
jugador2.join()

print("Ha ganado el jugador", resueltoPorJugador)
```

## Ejercicio 3: El juego de la Soga (4 puntos)
En el juego de la soga dos equipos tiran de una cuerda en direcciones opuestas para consegir desplazar el centro de la soga hacia 1 metro hasta su campo. En el torneo se juegan 14 partidas. El siguiénte código muestra cómo modelamos el juego mediante hilos. En cada instante, un solo participante o hilo puede desplazar hacia su campo la cuerda una distancia aleatoria de entre 0 y 10 cm.

```python
from threading import Thread, Lock
import random

#posición de la soga en cm
posiciónCentroSoga = 0
cerrojoCentro = Lock()

puntosEquipoIzquierda = 0
puntosEquipoDerecha = 0

threads = []

def tirar(dirección = "Izquierda"):
    while puntosEquipoIzquierda + puntosEquipoDerecha < 14:
        global posiciónCentroSoga
        # desplazamiento aleatorio de entre 0 y 10 cm
        desplazamiento = random.random() * 10
    
        # el signo del desplazamiento depende de la dirección
        if dirección == "Izquierda":
            desplazamiento *= -1
        
        with cerrojoCentro:
            posiciónCentroSoga += desplazamiento

# añadimos 10 participantes, 5 de cada equipo
for i in range(5):
    t1 = Thread(target=tirar, kwargs={"dirección": "Izquierda"})
    t2 = Thread(target=tirar, kwargs={"dirección": "Derecha"})
    t1.start()
    t2.start()
    threads.append(t1)
    threads.append(t2)
```
